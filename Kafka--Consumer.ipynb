```python
from kafka import KafkaConsumer
from json import loads
import json
from s3fs import S3FileSystem

consumer = KafkaConsumer(
    "demo_testing2",
    bootstrap_servers=["13.235.104.208:9092"],
    value_deserializer=lambda x: loads(x.decode("utf-8"))
)

s3 = S3FileSystem()
```

```python
for message in consumer:
    print(message.value)
    break
```

```python
import json

for count, message in enumerate(consumer):
    with s3.open(
        f"s3://mohd-subhan-stock-market-data/stock_market_{count}.json",
        "w"
    ) as file:
        json.dump(message.value, file)

    print(f"Uploaded stock_market_{count}.json")
```
